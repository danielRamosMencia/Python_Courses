<br>
Step 1 is create a list with our data, then the step 2 is use the pandas function "pd.DataFrame(args)". 
The argument is a list to create the dataframe, and also we can to personalize your headers
<br>

```python
l = [["Apple","Red"],["Banana", "Yellow"],["Prange", "Orange"]]
```

```python
data_frame = pd.DataFrame(l, columns=["Fruit","Color"])
data_frame
```
