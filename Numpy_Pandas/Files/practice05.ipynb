```python
data_frame = pd.read_csv("cereals.csv")
data_frame
```

<p>

```python
condition = data_frame['calories'] > 100
data_frame[condition]
```

<p>
Filtering using loc() function
</p>

```python
#1. To obtain as a print on screen
#data_frame.loc[0,'name']
#2. To obtain as a dataframe
data_frame.loc[[0],['name']]
```

```python
#1. To obtain a range of colors and rows
#data_frame.loc[0:4, 'name':'protein']
#2. To obtain a range of colors and a specific range of rows
#data_frame.loc[[5,8], 'name','protein']
#3. To obtain a specific range of colors and a specific range of columns 
data_frame.loc[[5,8],['name','protein']]
```

<p>
Filtering data using iloc() function
</p>

```python
data_frame.iloc[9, 2]
```

```python
#To get a data frame
data_frame.iloc[[9], [2]]
```

```python
data_frame.iloc[0:5, 0:3]
```

<p>
Adding and deleting rows and columns
</p>

```python
#adding a row
data_frame.loc[5] = ['Trix',110,1,25,27]
data_frame
```

```python
#deleting row
data_frame.drop(6, axis=0, inplace=True)
data_frame
```

```python
#Adding a column
data_frame['New_Column'] = ['A','B','C','D','E','F','G','H','I']
data_frame
```

```python
#deleting a column
data_frame.drop('New_Column', axis=1, inplace=True)
data_frame
```

<p>
Sort values to order by one condition
</p>

```python
data_frame.sort_values(by='calories')
```

```python
data_frame.sort_values(by='calories', ascending=False)
```
