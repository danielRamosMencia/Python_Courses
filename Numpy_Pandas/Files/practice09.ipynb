```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("iris.csv")
df
```

```python
x_axis = df.index
y_axis = df['petal.length']
sb.scatterplot(x=x_axis, y=y_axis, hue=df['variety'])
```

```python
sb.stripplot(x=df['variety'], y=df['petal.length'])
```

```python
plt.hist(df['petal.length'])
```

```python
sb.displot(df['petal.length'])
```

```python
sb.boxplot(x=df['petal.length'])
```

```python
sb.countplot(x=df['variety'])
```

```python
labels_list = ['SETOSA', 'VERSICOLOR', 'VIRGINICA']
plt.pie(df['variety'].value_counts(), labels=labels_list)
```
